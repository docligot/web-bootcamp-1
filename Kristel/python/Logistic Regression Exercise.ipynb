```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import math
```

```python
df = xl.parse("Sheet1")
```

```python
ind_var = df[["income", "channel", "age", "gender"]]
```

```python
ind_var
```

```python
channel_dummies = pd.get_dummies(ind_var["channel"], prefix="channel")
gender_dummies = pd.get_dummies(ind_var["gender"], prefix="gender")
```

```python
ind_var['age_group'] = np.where(ind_var.age < 25, 'Gen-Z', 
                               np.where(ind_var.age < 38, 'Gen-Y', 
                                        np.where(ind_var.age < 50, 'Gen-X', 
                                                'Boomer')))
```

```python
ind_var['income_group'] = np.where(ind_var.income < 20, '0-20', 
                          np.where(ind_var.income < 30, '21-30', 
                          np.where(ind_var.income < 40, '31-40',
                          np.where(ind_var.income < 50, '41-50',
                          np.where(ind_var.income < 60, '51-60',
                          np.where(ind_var.income < 70, '61-70',
                          np.where(ind_var.income < 80, '71-80',
                          np.where(ind_var.income < 90, '81-90', '91-100'))))))))
```

```python
ind_var = ind_var.drop('channel_group', axis=1)
```

```python
ind_var
```

```python
age_dummies = pd.get_dummies(ind_var["age_group"], prefix="age")
income_dummies = pd.get_dummies(ind_var["income_group"], prefix="income")
```

```python
income_dummies
```

```python
ind_var = ind_var.join(gender_dummies)
ind_var = ind_var.join(channel_dummies)
ind_var = ind_var.join(age_dummies)
ind_var = ind_var.join(income_dummies)
```

```python
ind_var_dummies = ind_var[[
 'gender_1', 
 'channel_2',
 'channel_3',
 'channel_4',
 'channel_5',
 'channel_6',
 'age_Boomer',
 'age_Gen-X',
 'age_Gen-Y',
 'income_21-30',
 'income_31-40',
 'income_41-50',
 'income_51-60',
 'income_61-70',
 'income_71-80',
 'income_81-90',
 'income_91-100']]
```

```python
ind_var_dummies.shape
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(ind_var_dummies, df.use)
```

```python
lg.score(ind_var_dummies, df.use)
```

```python
lg.predict(ind_var_dummies)
```

```python
output_df = df.join(pd.DataFrame(lg.predict(ind_var_dummies)))
output_df = output_df.rename(columns={0: 'pred_use'})
```

```python
output_df.shape
```

```python
sklearn.metrics.confusion_matrix(output_df.use, output_df.pred_use, labels=None, sample_weight=None)
```
