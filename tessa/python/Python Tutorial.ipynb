```python
# SK-LEARN TUTORIAL "LINEAR REGRESSION" 

# Load modules
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import math

# load excel source
xl = pd.ExcelFile(r"C:\Users\Dell\Desktop\Training Classes\UP OU Demo\Sales_train_test.xlsx") # loads file as binary
df = xl.parse("Sheet1") # parses sheet into dataframe

# create train and test sets
train = df.loc[df['set'] == 'Train']
test = df.loc[df['set'] == 'Test']
print(train)
print(test)

# initialize linear regression
from sklearn.linear_model import LinearRegression
lm = LinearRegression() # store linear regression as lm variable

# specify independent training variables
independent_variables = train.drop('sales_a', axis = 1)

# run the regression process
lm.fit (independent_variables, train.sales_a) # second term is the target variable

# output the intercept
print (lm.intercept_)

# output the coefficients
pd.DataFrame(list(zip(independent_variables.columns, lm.coef_)), columns = ['features', 'coefficients'])

# output the r-Squared
lm.score(independent_variables, train.sales_a)

# mean squared error
mseFull = math.sqrt(np.mean((train.sales_a - lm.predict(independent_variables)) ** 2))
print (mseFull)

# output the predicted target variables
lm.predict(independent_variables)

# do a scatter of the actual and predicted
plt.scatter(lm.predict(independent_variables), train.sales_a)
plt.xlabel("Predicted Sales")
plt.ylabel("Actual Sales")
plt.title("Predicted vs. Actual Sales")

# do the prediction on the test set
independent_variables_2 = test.drop('sales_a', axis = 1)
lm.predict(independent_variables_2)
lm.score(independent_variables_2, test.sales_a)
plt.scatter(lm.predict(independent_variables_2), test.sales_a)
plt.xlabel("Predicted Sales")
plt.ylabel("Actual Sales")
plt.title("Predicted vs. Actual Sales")
```
