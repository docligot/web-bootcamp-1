```python
# Import libraries and load dataset
from sklearn import preprocessing
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
lm = LinearRegression()

df = pd.read_csv(r"C:\Users\Cirrolytix\Documents\Python Bootcamp\linear_exercise.csv")
```

```python
df
```

```python
preprocessing.scale(df["Open Positions"])
```

```python
df["scale_positions"] = preprocessing.scale(df["Open Positions"])
df["leave_balance"] = preprocessing.scale(df["AvgLeave Balance"])
df["fte"] = preprocessing.scale(df["FTEs"])
```

```python
df["leave_per_fte"] = df["AvgLeave Balance"] / df["FTEs"]
```

```python
df
```

```python
region_df = pd.get_dummies(df["Region"])
dept_df = pd.get_dummies(df["Department"])
```

```python
df = df.join(region_df)
df = df.join(dept_df)
```

```python
df
```

```python
df.to_csv('modified_hr.csv')
```

```python
pred1 = df[["FTEs", "Open Positions", "AvgLeave Balance"]]
pred2 = df[["scale_positions", "leave_balance", "fte", "leave_per_fte"]]
pred3 = df[["scale_positions", "leave_balance", "fte", "leave_per_fte", "CALABARZON", "Cebu", "Davao"
]]
pred4 = df[["scale_positions", "leave_balance", "fte", "leave_per_fte", "CALABARZON", "Cebu", "Davao", "Finance", "HR", "IT"
]]

```

```python
pred4
```

```python
lm.fit(pred1, df["Overtime Hrs"])
lm.score(pred1, df["Overtime Hrs"])
```

```python
lm.fit(pred2, df["Overtime Hrs"])
lm.score(pred2, df["Overtime Hrs"])
```

```python
lm.fit(pred3, df["Overtime Hrs"])
lm.score(pred3, df["Overtime Hrs"])
```

```python
lm.fit(pred4, df["Overtime Hrs"])
lm.score(pred4, df["Overtime Hrs"])
```
